```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data = pd.read_csv("shrink_ray_dataset.csv")

shrinkage = data.drop("Power" , axis = 1) #independent variable
power = data["Power"] #dependent variable

# training data
shrinkage_train, shrinkage_test , power_train, power_test = train_test_split(shrinkage, power, test_size=0.2, random_state=42)


# linear regression model
reg = LinearRegression()

# training the model
reg.fit(shrinkage_train, power_train)

# get y values for drawing the line of linear regression
shrinkage_prediction = reg.predict(shrinkage_test)

# setup a figure
plt.figure(figsize = (20 , 10))
plt.title("linear regression")
plt.xlabel("shrinkage")
plt.ylabel("power")

plt.plot(shrinkage_test , shrinkage_prediction)
plt.scatter(shrinkage , power)



shrinkage_of_moon = pd.DataFrame( [85]  , columns = shrinkage.columns) #should be a pandas data frame for prediction

# get the power required to shrink the moon by predicting it using the linear regression
power_predicted = reg.predict(shrinkage_of_moon)[0]

print(f"to shrink the moon you need : {round(power_predicted , 3)} MW")

# scattering the point
plt.scatter(shrinkage_of_moon , power_predicted , color ='red')


plt.show()
```
